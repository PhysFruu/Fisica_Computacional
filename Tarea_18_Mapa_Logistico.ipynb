# Aplicación logística

$$x_{n+1}=1-rx_n^2$$
$x_0$ de -1 a 1,
$\mu=r$ de 0 a 2

````julia
#Función que evalua un punto inicial x0 con la serie y regresa el x final
function logistic_map(x0,r,n=1000)
    x=x0
    for i in 1:n
        x=1-r*x^2
    end
    return x
end
````

````julia
#A una lista de puntos iniciales, evalua uno por uno y regresa lista de valores finales xf
function Listxf_logistic_map(listx0,r)
    listxf=[]
        for i in 1:length(listx0)
        push!(listxf,logistic_map(listx0[i],r))
        end
    return listxf
end
````

````julia
using PyPlot
#Condiciones iniciales
listx0=linspace(-1,1,100)
#R's distintas
listr=linspace(0,2,1000)
#Evaluación y graficación para cada R
for i in 1:length(listr)
    #Se busca el vector de condicion inicial dada la mu
    listxf=Listxf_logistic_map(listx0,listr[i])
    #Se crea el Valor de R que corresponde a esa lista
    vecr=listr[i]*ones(length(listxf))
    #Se grafica R vs Xf
    scatter(vecr,listxf, marker=".")
end
````
